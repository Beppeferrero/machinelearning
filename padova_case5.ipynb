# PREDIZIONE PREZZI IMMOBILI 

```python
######### DATASET IMMOBILI DI PADOVA ###############
# Dataset minimo con solo 27 case e 8 attributi



# Librerie di lettura files
import pandas as pd
```

```python
# Librerie di machine learning
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
# Librerie di grafica
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Input del file Excel contenente il Dataset

nome_file=input("inserisci il nome del file excel (senza estensione xls) ")
```

```python
# Lettura del file Excel


dati = pd.read_excel('C:/users/mazza/Desktop/cartella_csv/'+nome_file+'.xls')
```

```python
# Features

X = dati[['zona','mq','stanze','anno_costruzione','anno_ristrutturazione','mq_garage','certificazione_energetica','riscaldamento']]
```

```python
# Target

y = dati['prezzo']
```

```python
# Controllo delle features
X
```

```python
# Controllo del target

y
```

```python
# Partitioning 

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

# Training

```python
######   ALGORITMO: REGRESSIONE LINEARE  ######

# Lancio dell'apprendimento con fit passando
# features di train e output di train

modello = LinearRegression()
modello.fit(X_train, y_train)
```

### Intercetta: b0

```python
# Calcolo e stampa dell'intercetta

print('intercetta', modello.intercept_)
```

```python
# Calcolo dei coefficienti

coefficienti = pd.DataFrame(modello.coef_, X.columns, columns = ['coefficienti'])
```

### Coefficienti: b1, b2,.....

```python
# Stampa dei coefficienti

print(coefficienti)
```

```python
y_calcolati = modello.predict(X)
```

# R^2

```python
######## VALUTAZIONE DEL MODELLO #################
# Calcolo del coefficiente di determinazione

print('R2', metrics.r2_score(y, y_calcolati))
```

```python
for r in range(0,8):
    if r==0:
        c1=coefficienti.iloc[r,0]
    elif r==1:
        c2=coefficienti.iloc[r,0]
    elif r==2:
        c3=coefficienti.iloc[r,0]
    elif r==3:
        c4=coefficienti.iloc[r,0]
    elif r==4:
        c5=coefficienti.iloc[r,0]        
    elif r==5:
        c6=coefficienti.iloc[r,0]   
    elif r==6:
        c7=coefficienti.iloc[r,0]    
    elif r==7:
        c8=coefficienti.iloc[r,0]            
print("c1= ",c1)
print("c2= ",c2)
print("c3= ",c3)
print("c4= ",c4)
print("c5= ",c5)
print("c6= ",c6)
print("c7= ",c7)
print("c8= ",c8)
```

```python
print("D A T I   I M M O B I L E")
continua = True
while continua:
    domanda=input('vuoi inserire i dati? ')
    if domanda.upper()=='SI':
        print("")
        zona=         int(input("             zona  "))
        mq=           int(input("      metri quadri "))
        stanze=       int(input("            stanze "))
        anno_costruzione =       int(input("  anno di costruz. "))               
        anno_ristrutturazione =       int(input(" anno di ristrutt. "))   
        mq_garage =        int(input("         mq garage "))   
        certificazione_energetica =    int(input("    certificazione ")) 
        riscaldamento=int(input("      riscaldamento")) 
      
        print("")
        prezzo=modello.intercept_+c1*zona+c2*mq+c3*stanze+c4*anno_costruzione+c5*anno_ristrutturazione+c6*mq_garage+c7*certificazione_energetica+c8*riscaldamento
        print("prezzo predetto",prezzo)
    else:
        break
```
