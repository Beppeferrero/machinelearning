```python
# ESERCIZIO CON MODELLO NON LINEARE (TERMINE AL QUADRATO E PRODOTTO DI 2 VARIABILI)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Creazione di 5 variabili x1, x2, x3, x4, x5
# distribuite secondo una gaussiana

x1 = np.random.normal(2,3,size=1000)   # x1 = sequenza (gaussiana) di 1000 valori centrata su 2, dev std = 3

# Distribuzione della variabile x1

plt.hist(x1)
plt.show()
```

```python
# Utilizziamo altre variabili

x2=np.random.normal(100,30,1000)
x3=np.random.normal(0,10,1000)
x4=np.random.normal(1,7,1000)
x5=np.random.normal(15,30,1000)
```

### Creazione array e Dataframe con i valori creati

```python
array = np.empty([1000, 5])    # Array di 1000 righe e 5 colonne vuote

array[:,0]=x1  # fa si che la sequenza x1 divenga la colonna 1 della matrice
array[:,1]=x2
array[:,2]=x3
array[:,3]=x4
array[:,4]=x5

# Trasformiamo la matrice (array) in un dataframe utilizzando pandas e la funzione DataFrame

df = pd.DataFrame(array)

# Diamo dei nomi alle colonne
df.columns=['col1','col2','col3','col4','col5']

# Visualizziamo il nostro dataframe
df
```

```python
# Vediamo come sono distribuite le variabili

sns.boxplot(data=df)
plt.show()
```

### Definizione della funzione

```python
# Definiamo una funzione che va a legare i dati creati con l'osservazione simulata
def funzione(a,b,c,d,e):
    return 0.1*x1 + 3*x2 + 0.1*x3**2 + 0.5*x4*x5

# la funzione genera delle osservazioni
target = funzione(x1, x2, x3, x4, x5)

# Aggiungiamo la colonna target
df['target'] = target

# Visualizzazione
df
```

```python
# La colonna target varia in funzione delle variabili col1, col2, col3, col4, col5
# Sono dati simulati (random) nella realtÃ  ci saranno dati reali,
# ma supponiamo che siano reali,
# inoltre supponiamo di non conoscere la relazione che c'Ã¨
# tra questi dati.
# La prima cosa che possiamo fare Ã¨ usare l'algoritmo
# piÃ¹ semplice che Ã¨ la REGRESSIONE LINEARE
```

### Creazione dell'algoritmo di ML - REGRESSIONE LINEARE

```python
#target = a*col1 + b*col2 + c*col3 + d*col4 + e*col5

# Relazione lineare tra ciÃ² che vogliamo andare
# a predire (target) e tutti gli input che abbiamo fornito come variabili
# Lo scopo Ã¨ trovare i valori di a,b,c,d,e per far si che 
# le nostre predizioni siano quanto piÃ¹ possibile vicine
# alla variabile target.  Quindi dobbiamo minimizzare una distanza,
# ad esempio (predizione - target)^2

#(predizione - target)**2

# Questa si chiama la minimizzazione degli scarti quadratici
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df.iloc[:,0:5]  # tutte le righe delle colonne da 0 a 5
y = df.iloc[:,5]  # variabile che vogliamo predire, tutte le righe della colonna 6

# Dividiamo i nostri dati in dati che utilizzeremo per l'apprendimento del ML
# e dati che utilizzeremo in fase di test; e questo viene fatto in
# automatico da scikit learn utilizzando la funzione train_test_split 
# a cui passiamo X, y (scegliendo una dimensione dei dati di test del 25%)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

# Lanciamo il nostro algoritmo di RL
lreg = LinearRegression()

# e questo lo fittiamo sui dati di training
lreg.fit(X_train,y_train) 

# A questo punto facciamo le predizioni utilizzando i dati di test
y_pred = lreg.predict(X_test)

# Ora vediamo se il modello funziona o no
# facendo uno scatter plot dei dati che abbiamo predetto rispetto ai dati di test
plt.scatter(y_pred, y_test)
plt.plot(y_pred, y_pred, color='red')
plt.show()
```

```python
# Se il modello fosse perfetto tutti i dati dovrebbero distribuirsi
# sulla linea rossa, vediamo invece che c'Ã¨ una certa dispersione 
# attorno alla linea rossa

# Valutiamo l'accuratezza del modello stampando lo score
score = lreg.score(X_test, y_test)
print('Lo score : ' +str(score))
```

```python
# Lo score ottenuto Ã¨ piuttosto basso
# Possiamo fare lo stesso sui dati di training

score = lreg.score(X_train, y_train)
print('Lo score : ' +str(score))
```
